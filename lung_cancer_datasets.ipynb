**First Step**:
Import padnas library, so we can read the file.

```python
import pandas as pd   # We have import Pandas library.

lung_cancer = pd.read_csv('/content/survey lung cancer.csv') # here we have read the file that we have download it from Kaggle website
```

**Second Step** **initinal exploration**:
After we read the file, we will navigate and explore it the data

```python
print(lung_cancer) # This line print all the details in the file
```

```python
print(lung_cancer.head(10)) # instead of printing all the data, this line just prints the first 10 lines
```

```python
print(lung_cancer.info()) # here we can see the parameters of the column, we have 14 integers (Discrete), and 2 Objects (categorical)
```

```python
print(lung_cancer.size) # This line of code shows the multiple of 309 rows * 16 columns
```

```python
print(lung_cancer.shape) # here we can see how many rows and columns
```

```python
print(lung_cancer.dtypes) # Sometimes we need to know the type of data we have so we can use the correct Visualisation table.
```

**Third Step Descriptive statistics**:
After we seen the data now we need to use the Descriptive statistics approach.

```python
# This line of code represents all the variables for the numerical and categorical
print(lung_cancer.describe(include='all')) # The numreical variables is (count, mean, std, min, 25%, 50%, 75%, max)
                                           # and for the categorical variables is (count, top, freq, unique)
```

```python
print(lung_cancer['ALCOHOL CONSUMING'].describe()) # here we can see only one variable instead of all the data
```

```python
# Let's print the range for the ['ALCOHOL CONSUMING']
max = lung_cancer['ALCOHOL CONSUMING'].max()  # here we create a variable called max and we assign to it the max value in the ['ALCOHOL CONSUMING']
min = lung_cancer['ALCOHOL CONSUMING'].min()  # here the lowest value
range = max - min
print('The maximum value:', max)
print('The lowest value:', min)
print('max - min = ', range)
```

**Forth step**: now we need to seprate the numerical and categorical data.

```python
# First, if we need to manipulate the data we need to import the Numpy library.
import numpy as np

numerical = lung_cancer.select_dtypes(include=['integer','float']).columns
print('The numerical parameters (integers), (float):\n ', numerical)
categorical = lung_cancer.select_dtypes(include=['object']).columns
print('The categorical parameters :\n', categorical)
```

**The fifth step the visualization**: Now we need to represent the data.

```python
# First we need to import the matplotlib library, so we can use it for shows the data

import matplotlib.pyplot as plt # here we import the library and give it name plt which is a common name

plt.hist(lung_cancer['AGE'],bins=15, color='lightblue', edgecolor='black') # The histogram is used with the continuous data (float), and because our data does not continue float is not fit for this type of representing
plt.xlabel('Age') # In this line, we named the X axis as 'Age'
plt.ylabel('Freq') # In this line, we named the Y axis as 'Freq'
plt.title('Age Distibution') # In this line, we named the header of histogram as 'Age Distibution'
plt.show() # Finally, show the histogram
```

```python
import seaborn as sns # here we use another library for the visualization
sns.histplot(lung_cancer['AGE'])
```

```python
# Now we will use the Box plot, and it's used with the numerical (cont), (disc)

plt.boxplot(lung_cancer['AGE'])
plt.show()
```

```python
# here we will use bat plot, and it's used with Categorical data

sns.barplot(lung_cancer['GENDER'])
```

```python
# Here we will use Pie chart:
# Catergoical data

lung_cancer['GENDER'].value_counts().plot(kind='pie', autopct='%1.2f%%', explode=[0.1,0.01], shadow=True)
plt.show()
```

```python
# Here we will use Scatter plot
sns.regplot(x = lung_cancer['AGE'], y = lung_cancer['YELLOW_FINGERS'], data=lung_cancer)
```

```python
# We can use another visualisation way to show all the data which is pairplot

sns.pairplot(lung_cancer) # as you can see we print all the possible relation between the data prarameters
```

```python
# here we will print the relation between one Categrical and one numerical

sns.boxplot(x = lung_cancer['GENDER'], y = lung_cancer['AGE'])
```

```python
# Here we will print the relation between two Cat and one num

sns.boxplot(x = lung_cancer['GENDER'], y = lung_cancer['LUNG_CANCER'], hue=lung_cancer['AGE'], data=lung_cancer)
```

```python
# Here we will use the heatmap to show all the relations

plt.figure(figsize=(12, 7))
sns.heatmap(lung_cancer.corr(), annot=True, vmin = -1, vmax = 1)
plt.show()
```
